```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Dataset/Data.csv")

# Data Cleaning

print("First few rows of the dataset:")
print(df.head())

print("\nMissing values in the dataset:")
print(df.isnull().sum())

df['datetime'] = pd.to_datetime(df['datetime'])
```

## Line plot of 'close' over time

```python
plt.figure(figsize=(12, 6))
plt.plot(df['datetime'], df['close'], marker='o', linestyle='-')
plt.title('Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()
```

## Box plot of 'volume'

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['volume'])
plt.title('Volume Distribution')
plt.ylabel('Volume')
plt.grid(True)
plt.show()
```

## Histogram of 'Fear_and_Greed_Index'

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Fear_and_Greed_Index'], bins=20, color='skyblue')
plt.title('Fear and Greed Index Distribution')
plt.xlabel('Fear and Greed Index')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## Scatter Plot

```python
date = pd.to_datetime(df['datetime'])
btc_close = df['close']
buy_signal = df['signal'] == 'buy'
sell_signal = df['signal'] == 'sell'

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(date, btc_close, label='BTC Close Price')

ax.scatter(date[buy_signal], btc_close[buy_signal], marker='^', color='green', label='Buy Signal')
ax.scatter(date[sell_signal], btc_close[sell_signal], marker='v', color='red', label='Sell Signal')

ax.set_title('BTC Price with Predicted Buy/Sell Signals')
ax.set_xlabel('Date')
ax.set_ylabel('BTC Close Price')

ax.legend()

plt.xticks(rotation=45)
plt.show()
```
